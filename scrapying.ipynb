## 1) Scrap Autoprix

```python
import requests 
from bs4 import BeautifulSoup

filename = "autoprix_data.csv"
f = open(filename,"w")
headers = "Voiture|Prix|Annee|Kilometrage|Carburant\n"
f.write(headers)

for i in range(2,4):
    
    url='https://www.autoprix.tn/recherche?cp=' +str(i)
    response=requests.get(url)
    soup=BeautifulSoup(response.text,features="html.parser")
        #n5dmou 3ala classe mo3ayen: grabs each products

    page_soup=soup.findAll("div",{"class":"col-sm-6 col-md-6 col-12"})
    for container in page_soup:
        Voiture = container.find("h1",{"class":"font-weight-black title"})
        Prix = container.find("h3",{"class":"title font-weight-black"})
        Container=container.find("div",{"class":"row ma-0 mt-1 card-detail text-center no-gutters"})
        Container=str(Container.text).split(' ')
        if (str(Container[0])!="NC"):
            print(Voiture.text + "|" + Prix.text + "|" +str(Container[0]) + "|" + str(Container[1])+ "|" + str (Container[-1]))
            
            
            
        f.write(Voiture.text + "|" + Prix.text + "|" +str(Container[0]) + "|" + str(Container[1])+ "|" + str (Container[-1]) + "\n" )
        
        
        
f.close()
            
```

## 2) Scrap Automobile

```python
import requests 
from bs4 import BeautifulSoup

filename = "Automobile2.csv"
f = open(filename,"a")
headers = "Voiture|Prix|Annee|Kilometrage|Localite|Carburant|Vitesse\n"
f.write(headers)

for i in range(1,3):
    
    url='https://www.automobile.tn/fr/occasion/2' #+str(i)
    response=requests.get(url)
    soup=BeautifulSoup(response.text,features="html.parser")

    
    page_soup=soup.findAll("div",{"class":"occasion-item"})
    for container in page_soup:
        Container=container.find("div",{"class":"thumb-caption"})
        Container=str(Container.text).split('                    ')
        Voiture  = str(Container[0])
        Prix = str(Container[1])
        Container1=container.find("ul",{"class":"specs-preview"})
        Container1=str(Container1.text).split('\n')
        Annee = str(Container1[3])
        Kilometrage = str(Container1[7])
        Localite = str(Container1[11])
        Carburant = str(Container1[15])
        Vitesse = str(Container1[19])
        print(Voiture + "|" + Prix + "|" + Annee + "|" + Kilometrage + "|" + Localite + "|" + Carburant + "|" + Vitesse + "\n" )
        f.write(Voiture + "|" + Prix + "|" + Annee + "|" + Kilometrage + "|" + Localite + "|" + Carburant + "|" + Vitesse + "\n" )

f.close()
            
```
